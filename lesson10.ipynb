# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod2.csv"
df = pd.read_csv(filepath, header=0)
```

```python
# Visualisations
# CPU frequency is postive against price
# Screen size is unsure, possible negtative against price
# Weight is also unsure, possible negative against pric

sns.regplot(x = "Weight_pounds", y = "Price", data = df)
plt.ylim(0,)
```

```python
# Correlations
# Nothing too strong except Screen size and Weight being 0.8 which makes sense, nothing strong against price except 0.36 on CPU frequency.

df[["CPU_frequency", "Screen_Size_inch", "Weight_pounds", "Price"]].corr()
```

```python
# Visualising catogorical values

sns.boxplot(x = "CPU_core", y = "Price", data = df)
```

```python
# Descriptive statistics

print(df.describe(include = ["object"]))
```

```python
# Groupby and PivotTables

df_group_one = df[["GPU", "CPU_core", "Price"]]
df_grouped = df_group_one.groupby(["GPU", "CPU_core"], as_index = False).mean()
print(df_grouped)
```

```python
df_pivot = df_grouped.pivot(index = "GPU", columns = "CPU_core")
print(df_pivot) 
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(df_pivot, cmap='RdBu')

#label names
row_labels = df_pivot.columns.levels[1]
col_labels = df_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

fig.colorbar(im)
```

```python
# Pearson's Coefficient

for param in ['RAM_GB','CPU_frequency','Storage_GB_SSD','Screen_Size_inch','Weight_pounds','CPU_core','OS','GPU','Category']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['Price'])
    print(param)
    print("The Pearson Correlation Coefficient for ",param," is", pearson_coef, " with a P-value of P =", p_value)
```
