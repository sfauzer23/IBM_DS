```python
import pandas as pd
import numpy as np
```

```python
# Import data

filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(filepath, header=None)
```

```python
# Set Columns

df.columns = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
# Replace the question mark sign with NaN

df1 = df.replace('?',np.NaN)
```

```python
# This will drop the rows with missing values in the price column

df1 = df1.dropna(subset=["price"], axis=0)
```

```python
# Save dataset as a csv file

df.to_csv("automobile.csv", index=False)
```
