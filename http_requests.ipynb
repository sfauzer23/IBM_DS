Requests is a Python Library that allows you to send HTTP/1.1 requests easily. We can import the library as follows:

```python
import requests
```

```python
import os 
from PIL import Image
from IPython.display import IFrame
```

```python
url='https://www.ibm.com/'
r=requests.get(url) # Get request, sends the responce of 200 which is okay
r.status_code
```

```python
print(r.request.headers)
print("request body:", r.request.body) # None as its a get
```

```python
header=r.headers
print(r.headers)
```

```python
header['Date'] # Can return values like this
```

```python
url='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/IDSNlogo.png'
```

```python
r=requests.get(url) # Shows that it is an image in the Content-Type
print(r.headers)
```

```python
url_get='http://httpbin.org/get'
payload={"name":"Joseph","ID":"123"}
r=requests.get(url_get,params=payload)
```

```python
print(r.text)
r.headers['Content-Type']
```
